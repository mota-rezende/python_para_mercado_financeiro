Realizando a extração de múltiplos ativos

```python
import pandas as pd
```

```python
ativos = ["PETR4", "VALE3", "ABEV3", "WEGE3", "EMBR3", "RADL3", "CVCB3", "CIEL3", "OIBR3", "KNRI11"]
```

```python
inicio = "2020-01-01"
fim = "2021-08-15"
```

```python
df = pd.DataFrame()

for i in ativos:
    ticker = str(i) + ".SA"
    df[i] = yf.download(ticker, start = inicio, end = fim)["Adj Close"]
```

```python
df.head()
```

```python
df.plot(figsize = (10, 10));
```

### Construção de carteira e comparação com o IBOV

```python
def retorna_ativos(ativos, inicio, fim):
    df = pd.DataFrame()

    for i in ativos:
        ticker = str(i) + ".SA"
        df[i] = yf.download(ticker, start = inicio, end = fim)["Adj Close"]

    return df
```

```python
carteira = ["PETR4", "VALE3", "ABEV3", "WEGE3", "EMBR3", "RADL3", "CVCB3", "CIEL3", "OIBR3", "KNRI11"]
```

```python
precos = retorna_ativos(carteira, inicio, fim)
```

```python
precos.head()
```

#### Quantos papéis compramos de cada um desses ativos?

```python
primeira = precos.iloc[0]
```

```python
primeira
```

```python
comprados = round(1000/primeira)
```

```python
comprados
```

#### Calcular o patrimônio liquido

```python
PL = precos * comprados
```

```python
PL.head()
```

```python
PL.tail()
```

```python
PL["Minha carteira"] = PL.iloc[:].sum(axis = 1)
```

```python
PL.head()
```

```python
PL.tail()
```

### Baixando os dados do IBOV para fazer comparação

```python
ibov = yf.download("^BVSP", start = inicio, end = fim)
```

```python
ibov.head()
```

```python
ibov.rename(columns = {'Adj Close': 'IBOV'}, inplace = True)
```

```python
ibov = ibov.drop(ibov.columns[[0, 1, 2, 3, 5]], axis = 1)
```

```python
ibov.head()
```

```python
PL.index = pd.to_datetime(PL.index)
```

```python
novo_df = pd.merge(ibov, PL, how = 'inner', on = 'Date')
novo_df.head()
```

```python
novo_normalizado = novo_df/novo_df.iloc[0]
novo_normalizado.head()
```

```python
novo_normalizado.tail()
```

```python
novo_normalizado[['IBOV', 'Minha carteira']].plot(figsize = (10, 10));
```
