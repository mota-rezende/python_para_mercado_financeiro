```python
import yfinance as yf
```

```python
msft = yf.download('MSFT', start = '2021-01-01')
```

```python
msft.head()
```

```python
msft_ohlc = msft[[ 'Open', 'High', 'Low', 'Close']]
```

```python
msft_ohlc.head()
```

```python
msft_ohlc.plot();
```

```python
msft_ohlc['Close'].plot();
```

```python
abev = yf.download('ABEV3.SA', start = '2021-01-01')
```

```python
abev.head()
```

```python
abev_ohlc = abev[['Open', 'High', 'Low', 'Close']]
```

```python
abev_ohlc.head()
```

### Exportando o arquivo como tabela .csv

```python
abev_ohlc.to_csv('ABEV3_OHLC.csv', sep = ';', decimal = ',')
```

### Importando o arquivo para o Python

```python
import pandas as pd
```

```python
dados = pd.read_csv('ABEV3_OHLC.csv', sep = ';', decimal = ',')
```

```python
dados.plot();
```

```python
dados.index
```

```python
dados['Date'] = pd.to_datetime(dados['Date'])
```

```python
dados.index = dados['Date']
```

```python
dados.head()
```

```python
dados.head()
```

```python
type(dados.index)
```

```python
dados.drop(['Date'], axis = 1, inplace = True)
```

```python
dados.head()
```

```python
dados.plot();
```
